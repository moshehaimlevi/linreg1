```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


advertising_investment = np.array([10_000, 15_000, 20_000, 25_000, 30_000, 35_000, 40_000, 45_000, 50_000]).reshape(-1, 1)


sales_growth = np.array([25_000, 30_000, 40_000, 45_000, 50_000, 60_000, 65_000, 70_000, 80_000])

model = LinearRegression()
model.fit(advertising_investment, sales_growth)

# Display model parameters
print(f"Slope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")
equation = f"y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}"
print(f"Line equation: {equation}")

# Prediction investment to get 100k in sales:
target_growth = 100_000
amount_needed = (target_growth - model.intercept_) / model.coef_[0]
print(f"To get a revenue of 100,000, approximately {amount_needed:.2f} needed")

# investing 60k
investment = 60_000
predicted_growth = model.predict([[investment]])
print(f"Predicted sales growth for {investment:,} investment: {predicted_growth[0]:,.2f}")

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(advertising_investment, sales_growth, color='blue', label='Data points')
plt.plot(advertising_investment, model.predict(advertising_investment), color='red', label='Regression line')
plt.scatter([[amount_needed]], [target_growth], color='green', s=100, label='Prediction for 100k')

plt.title('Linear Regression - Advertising Investment vs. Sales Growth')
plt.xlabel('Advertising Investment')
plt.ylabel('Sales Growth')
plt.grid(True)
plt.legend()

# Proper placement for equation text
plt.text(15_000, 30_000, equation, fontsize=12, color='black')

plt.show()
```
